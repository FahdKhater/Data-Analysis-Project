```python
import pandas as pd
url = 'https://statranker.org/cities-urban-life/top-100-cities-by-international-tourist-arrivals-2025/'
df = pd.read_html(url)
```

```python
import pandas as pd

df_combined = pd.concat(df, ignore_index=True)

target_cities = [
    "Paris", "Barcelona", "Tokyo", "New York", "London",
    "Rome", "Amsterdam", "Sydney", "Bangkok", "Istanbul",
    "Cairo", "Rio de Janeiro", "Venice", "Los Angeles"
]
df_combined['City'] = df_combined['City, Country'].apply(lambda x: x.split(",")[0].strip())
df_filtered = df_combined[df_combined['City'].isin(target_cities)].copy()

def match_city(city):
    for target in target_cities:
        if target.lower() in city.lower():
            return target
    return None

df_combined['City_Standard'] = df_combined['City'].apply(match_city)
df_filtered = df_combined[df_combined['City_Standard'].notna()].copy()

df_filtered['International Tourist Arrivals (2025)'] = (
    df_filtered['International Tourist Arrivals (2025)']
    .str.replace('≈','')
    .str.replace('M','')
    .str.strip()
    .astype(float)
)

```

```python
df_filtered = df_filtered[['City_Standard', 'International Tourist Arrivals (2025)']]
df_filtered.columns = ["City", "International Tourist Arrivals (Millions)"]

df_filtered.reset_index(drop=True, inplace=True)
```

```python
df_filtered = df_filtered.drop_duplicates(subset='City', keep='first').reset_index(drop=True)
```

```python
df_filtered.to_csv("TouristArrivals.csv", index=False)

```
