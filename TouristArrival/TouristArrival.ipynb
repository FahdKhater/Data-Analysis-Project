```python
import pandas as pd
url = 'https://statranker.org/cities-urban-life/top-100-cities-by-international-tourist-arrivals-2025/'
df = pd.read_html(url)
```

```python
import pandas as pd

df_combined = pd.concat(df, ignore_index=True)

target_cities = [
    "Paris", "Barcelona", "Tokyo", "New York", "London",
    "Rome", "Amsterdam", "Sydney", "Bangkok", "Istanbul",
    "Cairo", "Rio de Janeiro", "Venice", "Los Angeles"
]
df_combined['City'] = df_combined['City, Country'].apply(lambda x: x.split(",")[0].strip())

df_filtered = df_combined[df_combined['City'].isin(target_cities)].copy()

df_filtered['International Tourist Arrivals (2025)'] = (
    df_filtered['International Tourist Arrivals (2025)']
    .str.replace('≈','')
    .str.replace('M','')
    .str.strip()
    .astype(float)
)

df_filtered.reset_index(drop=True, inplace=True)

print(df_filtered[['City', 'International Tourist Arrivals (2025)']])
```
