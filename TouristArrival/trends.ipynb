```python
from pytrends.request import TrendReq
import pandas as pd

cities = ["Paris", "Barcelona", "Tokyo", "New York", "London", 
          "Rome", "Amsterdam", "Sydney", "Bangkok", "Istanbul", 
          "Cairo", "Rio de Janeiro", "Venice", "Los Angeles"]

pytrends = TrendReq()
df_all = pd.DataFrame()
for city in cities:
    kw = [f"Visit {city}"]
    pytrends.build_payload(kw, timeframe='2024-01-01 2025-12-31')
    df = pytrends.interest_over_time()
    df = df.drop(columns=['isPartial'])
    df[city] = df[f"Visit {city}"] / df[f"Visit {city}"].max()
    df_all[city] = df[city]
```

```python
df_all.to_csv("trends.csv")
```
